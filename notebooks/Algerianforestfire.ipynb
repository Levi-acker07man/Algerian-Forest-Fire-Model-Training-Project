```python
import pandas as pd
import matplotlib.pyplot as  plt
import numpy as np
%matplotlib inline
```

```python
dataset=pd.read_csv("Algerian_forest_fires_dataset.csv")
```

```python
dataset.head()
```

```python
##Datacleaning process
##mising values
dataset[dataset.isnull().any(axis=1)]
```

```python
#making columns
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
##Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.head()
```

```python
df.columns
```

```python
##Change the required columns as integer datatype
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
##Changing the other columns to float datatype
cat_features = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
# 1. Strip any hidden spaces from the columns first
df.columns = df.columns.str.strip()

# 2. List all the columns that should be numbers
num_columns = ['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Temperature', 'RH', 'Ws']

# 3. Force them to numbers. Rogue text will become NaN automatically.
for col in num_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 4. Drop the rows that contained the rogue text or typos
df = df.dropna().reset_index(drop=True)
```
